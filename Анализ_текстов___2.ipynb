Архитектура **трансформера** (Transformer) — это нейросетевая модель, представленная в 2017 году в статье *«Attention Is All You Need»*. Её ключевая особенность — использование **механизма внимания (attention mechanism)** вместо традиционных рекуррентных (RNN) или сверточных (CNN) слоёв. Это позволяет модели эффективно обрабатывать длинные последовательности и улавливать глобальные зависимости между словами.

---

### **Основные компоненты трансформера**
1. **Энкодер (Encoder)**:
   - Преобразует входные данные (например, текст) в векторные представления.
   - Состоит из идентичных слоёв, каждый из которых включает:
     - **Self-Attention**: Анализирует связи между всеми словами в последовательности.
     - **Feed Forward Neural Network**: Обрабатывает данные независимо для каждого слова.

2. **Декодер (Decoder)**:
   - Генерирует выходную последовательность (например, перевод или ответ).
   - Также содержит слои с Self-Attention и Feed Forward, но добавляет **Masked Attention**, чтобы учитывать только предыдущие слова при генерации.

3. **Позиционные эмбеддинги**:
   - Поскольку трансформеры не имеют встроенного понятия порядка, позиции слов кодируются специальными векторами.

---

### **Механизм внимания (Self-Attention)**
Self-Attention позволяет модели определять, насколько важно каждое слово в последовательности относительно других. Работает в три шага:

1. **Создание векторов Query, Key, Value**:
   - Каждое слово преобразуется в три вектора:
     - **Query** (запрос) — что ищем.
     - **Key** (ключ) — с чем сравниваем.
     - **Value** (значение) — информация, которую извлекаем.
   - Получаются умножением входных эмбеддингов на обучаемые матрицы \( W^Q, W^K, W^V \).

2. **Расчет весов внимания**:
   - Для каждого слова вычисляется **внимание** к другим словам через скалярное произведение Query и Key:
     \[
     \text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V,
     \]
     где \( d_k \) — размерность ключей (для стабилизации градиентов).

3. **Объединение результатов**:
   - Веса внимания умножаются на векторы Value, чтобы получить итоговое представление слова.

---

### **Пример работы Self-Attention**
Рассмотрим предложение:  
**«The cat didn’t cross the street because it was too tired»**  

Задача: Определить, к чему относится слово **«it»**.

1. **Шаг 1**: Для слова **«it»** создаются векторы Query.
2. **Шаг 2**: Query **«it»** сравнивается с Key всех слов в предложении.  
   - Наибольшее сходство будет с Key слова **«cat»** (т.к. «it» ссылается на «cat»).
3. **Шаг 3**: Веса внимания умножаются на Value соответствующих слов.  
   - Вектор **«it»** обогащается информацией из **«cat»**.

**Визуализация весов внимания**:

| Слово  | Вес внимания для «it» |
|--------|-----------------------|
| cat    | 0.9                   |
| street | 0.05                  |
| tired  | 0.03                  |
| ...    | ...                   |

---

### **Multi-Head Attention**
Для улавливания разных типов зависимостей трансформер использует **несколько голов внимания**:
- Каждая «голова» обучается фокусироваться на разных аспектах (например, синтаксис, семантика).
- Результаты всех голов объединяются в итоговый вектор.

---

### **Преимущества трансформеров**
1. **Параллелизация**: Обработка всех слов одновременно (в отличие от RNN).
2. **Глобальный контекст**: Учитывает связи между любыми словами, даже на большом расстоянии.
3. **Масштабируемость**: Легко увеличивать размер модели (например, GPT-3).

---

### **Пример архитектуры (кодировщик)**
1. Входные эмбеддинги + позиционные кодировки.
2. **Слой 1**:
   - Multi-Head Self-Attention.
   - Residual connection + LayerNorm.
   - Feed Forward Network.
3. Повторение шага 2 для N слоёв (например, 12 слоёв в BERT).

---

Трансформеры стали основой для современных моделей (GPT, BERT, T5), достигших прорыва в NLP. Их способность анализировать сложные зависимости в данных делает их универсальными для задач перевода, генерации текста и классификации.

Конвертация форматов
Часто данные приходится конвертировать между форматами. Например, из IOB в формат spaCy:

```python
def convert_iob_to_spacy(iob_labels, tokens):
    entities = []
    current_label = None
    start = 0
    for i, (token, label) in enumerate(zip(tokens, iob_labels)):
        if label.startswith("B-"):
            if current_label is not None:
                entities.append((start, i-1, current_label))
            current_label = label[2:]
            start = i
        elif label == "O" and current_label is not None:
            entities.append((start, i-1, current_label))
            current_label = None
    return entities
```

Вопрос: Для анализа раздела проектной документации на соответствие нормативным требованиям на первом этапе эффективней использовать правила для определения наименований разделов, далее для выделения наименования нормативной документации из раздела и сравнения его с перечнем приложенный к заданию на проектирование. Я правильно понимаю общий подход п анализу?

пользователь спрашивает, правильно ли он понимает общий подход к анализу проектной документации на соответствие нормативным требованиям. Он предлагает использовать сначала правила для определения наименований разделов, затем выделить названия нормативных документов и сравнить их с перечнем из задания на проектирование. Нужно проверить, верен ли такой подход.

Сначала вспомню, как обычно проводится анализ соответствия. Часто начинают с структурирования документации, чтобы понять, где искать нужную информацию.
```python
# 1.1 Предварительная обработка и проверка
# ----------------------
class RuleBasedChecker:   # "Проверка на основе правил"
    def __init__(self):
        # Словарь паттернов с читаемыми названиями разделов
        self.section_patterns = {       # "мои. разделы патернов"
            "Технологическая схема": r'(?i)технологическая\s+схема',
            "Потребности ресурсов": r'(?i)потребности\s+ресурсов',
            # ... остальные разделы
        }

        # Предкомпилированные регулярные выражения
        self.compiled_sections = {
            name: re.compile(pattern) for name, pattern in self.section_patterns.items()
        }
        self.norm_regex = re.compile(r'\b(?:ГОСТ|СП|СНиП)\s*[\d.-]+')

    def find_missing_sections(self, text):
        """Возвращает читаемые названия отсутствующих разделов."""
        if not isinstance(text, str):
            raise TypeError("Текст должен быть строкой")

        missing = []
        for name, regex in self.compiled_sections.items():
            if not regex.search(text):
                missing.append(name)
        return missing

    def extract_norms(self, text):
        """Извлекает нормативные ссылки из текста."""
        if not isinstance(text, str):                        # проверка типа входных данных
            raise TypeError("Текст должен быть строкой")

        return self.norm_regex.findall(text)
```

**Четкое разделение логики и представления**

Паттерны хранятся вместе с их человеко-читаемыми названиями, но логика работы использует только предкомпилированные объекты.

**Производительность**

Регулярные выражения компилируются однократно при создании экземпляра класса, а не при каждом вызове методов.

**Удобство использования**

Методы возвращают понятные названия разделов вместо технических паттернов:

Возвращает: ['Потребности ресурсов'] вместо ['потребности\\s+ресурсов'] *Курсив*

**Стабильность**

Проверки типа входных данных предотвращают ошибки выполнения при некорректных аргументах.

**Масштабируемость**

Добавление новых проверок требует простого расширения словаря section_patterns без изменения логики методов.

```python
# 2. ML модель (Fine-tuned RuBERT)
# ----------------------
class DocumentClassifier:
    def __init__(self):
        self.tokenizer = BertTokenizer.from_pretrained("DeepPavlov/rubert-base-cased")
        self.model = BertForSequenceClassification.from_pretrained("path_to_your_fine_tuned_model")

    def predict(self, text):
        inputs = self.tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
        outputs = self.model(**inputs)
        return np.argmax(outputs.logits.detach().numpy())

# ----------------------
# 3. Интеграция с внешними базами нормативов
# ----------------------
class NormativeDBClient:
    def __init__(self):
        self.api_url = "https://normative-db.example.com/check"

    def check_norm_status(self, norm_code):
        try:
            response = requests.get(f"{self.api_url}?code={norm_code}", timeout=3)
            return response.json()['status']
        except:
            return "Не удалось проверить"

# ----------------------
# 4. Постобработка и рекомендации
# ----------------------
class ReportGenerator:
    @staticmethod
    def generate(report_data):
        recommendations = []

        # Проверка разделов
        if report_data['missing_sections']:
            recommendations.append(f"Отсутствуют обязательные разделы: {', '.join(report_data['missing_sections']}")

        # Проверка нормативов
        for norm, status in report_data['norms_status'].items():
            if status != "актуальный":
                recommendations.append(f"Норматив {norm} требует обновления ({status})")

        return {
            "compliance_level": "Соответствует" if not recommendations else "Частично соответствует",
            "recommendations": recommendations
        }

# ----------------------
# Основной рабочий процесс
# ----------------------
def analyze_document(document_text):
    # Инициализация компонентов
    rule_checker = RuleBasedChecker()
    ml_classifier = DocumentClassifier()
    db_client = NormativeDBClient()

    # Анализ по правилам
    missing_sections = rule_checker.find_missing_sections(document_text)
    detected_norms = rule_checker.extract_norms(document_text)

    # ML анализ
    ml_compliance = ml_classifier.predict(document_text)

    # Проверка нормативов в базе
    norms_status = {}
    for norm in detected_norms:
        norms_status[norm] = db_client.check_norm_status(norm)

    # Формирование отчета
    report_data = {
        "missing_sections": missing_sections,
        "norms_status": norms_status,
        "ml_verdict": "Соответствует" if ml_compliance == 1 else "Требует проверки"
    }

    return ReportGenerator.generate(report_data)

# ----------------------
# Пример использования
# ----------------------
if __name__ == "__main__":
    sample_text = """
    Раздел 1. Пояснительная записка. Соответствует ГОСТ Р 12345-2001.
    Раздел 3. Смета. Использован устаревший СНиП 12-34-1985.
    """

    report = analyze_document(sample_text)
    print("Статус соответствия:", report['compliance_level'])
    print("Рекомендации:")
    for rec in report['recommendations']:
        print("- ", rec)
```

```python
import pdfplumber

#  Анализ типов содержимого в PDF:
def analyze_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            print(f"Страница {i+1}:")
            print(f"- Текстовых объектов: {len(page.chars)}")
            print(f"- Векторных путей: {len(page.curves)}")
            print(f"- Растровых изображений: {len(page.images)}")
            print(f"- Областей с растровым фоном: {len(page.get_image_bboxes())}")
```
